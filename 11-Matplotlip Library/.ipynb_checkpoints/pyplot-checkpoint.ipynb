```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#تمرین1
N_A_X = np.arange(1,9)
```

```python
N_A_X 
```

```python
N_A_Y = np.arange(3,27,3)
```

```python
N_A_Y
```

```python
plt.xlabel("N_A_X")
plt.ylabel("N_A_Y")
plt.title("PLOT")
plt.plot(N_A_X, N_A_Y, color = "orange")
plt.show()
```

```python
#تمرین 2
LS_X = np.linspace(-4,4,50)
LS_Y = LS_X ** 2
plt.plot(LS_X, LS_Y, "r.")
plt.show()
```

```python
#تمرین 3
plt.plot(LS_X, np.sin(LS_X), "r.")
plt.plot(LS_X, np.cos(LS_X), "g--")
plt.show()
```

```python
# تمرین 4
N_A_X = np.arange(1,9)
N_A_Y = np.arange(3,27,3)
plt.bar(N_A_X, N_A_Y)
plt.show()
```

```python
#تمرین 5
plt.polar(N_A_X, N_A_Y)
plt.show()
```

```python
#تمرین 6
plt.step(N_A_X, N_A_Y)
plt.show()
```
