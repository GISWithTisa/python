```python
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
#تمرین1
X = np.arange(0,math.pi*2,0.05)
Figure = plt.figure()
Axes = Figure.add_axes([0,0,1,1])
Y = np.sin(X)
Axes.plot(X,Y)
Axes.set_xlabel("angle")
Axes.set_title("sine")
Axes.set_xticks([0,2,4,6])
Axes.set_xticklabels(["zero","two","four","six"])
Axes.set_yticks([-1,0,1])
plt.show()
```

```python
#تمرین2
X = np.arange(0,math.pi*2,0.05)
Figure = plt.figure()
Axes = Figure.add_axes([0,0,1,1])
Y = np.sin(X)
Axes.plot(X,Y)
Axes.set_xlabel("angle")
Axes.set_title("sine")
Axes.set_xticks([0,2,4,6])
Axes.set_xticklabels(["zero","two","four","six"])
Axes.set_yticks([-1,0,1])
Axes.spines["right"].set_visible(False)
Axes.spines["top"].set_visible(False)
plt.show()
```

```python
#تمرین3
X = np.arange(0,math.pi*2,0.05)
Figure = plt.figure()
Axes = Figure.add_axes([0,0,1,1])
Y = np.sin(X)
Axes.plot(X,Y)
Axes.set_xlabel("angle")
Axes.set_title("sine")
Axes.set_xticks([0,2,4,6])
Axes.set_xticklabels(["zero","two","four","six"])
Axes.set_yticks([-1,0,1])
Axes.spines["right"].set_visible(False)
Axes.spines["top"].set_visible(False)
Axes.grid(color="b", ls= "-", lw = 0.25)
plt.show()
```
