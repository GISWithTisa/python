```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#این کد با نوت بوک ژوپیتر ساخته شده و به صورت اجرا شده اضافه اش کردیم. دست بهش نزن بذار همینطوری بمونه
Figure = plt.figure(facecolor = 'red')
X = np.arange(-1,5,0.1)
Y = np.exp(-X)*np.cos(5*X)
plt.plot(X,Y)
plt.show()
```
