```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#تمرین2
sns.set_style('whitegrid')
plt.figure()
X1 = [10,20,5,40,0]
X2 = [30,43,9,7,20]
plt.plot(X1,label = "Group A")
plt.plot(X2,label = "Group B")
plt.legend()
plt.show()
```

```python
#تمرین3
sns.set()
plt.figure()
X1 = [10,20,5,40,0]
X2 = [30,43,9,7,20]
with sns.axes_style('dark'):
    plt.plot(X1,label = "Group A")
    plt.plot(X2,label = "Group B")
plt.legend()
plt.show()
```

```python
#تمرین4
sns.set_context("poster")
plt.figure()
X1 = [10,20,5,40,0]
X2 = [30,43,9,7,20] 
plt.plot(X1,label = "Group A")
plt.plot(X2,label = "Group B")
plt.legend()
plt.show()
```
